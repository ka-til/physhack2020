```python
import numpy as np
import matplotlib.pylab as plt
```

```python
wavenumberO2 = np.loadtxt('wavenumber_atm_O2.txt.csv', delimiter=',')
transmittanceO2 = np.loadtxt('transmittance_spectrum_atm_O2.txt.csv', delimiter=',')

%matplotlib notebook
plt.plot(wavenumberO2, np.log(1/transmittanceO2))
plt.title('Oxygen Absorption spectrum')
plt.xlabel('Wavenumber[cm^-1]', fontsize = 18)
plt.ylabel('Absorbance', fontsize = 18)
```
